```python
# Cria uma cópia modificada excluindo dados faltantes e salva.
df_clean = df.dropna()
```

```python
# Remover dados duplicados e salva uma cópia.
df_clean_noduplicate = df_clean.drop_duplicates()

df_clean_noduplicate.info()
```

```python
# Axis=0 (linha) / Axis=1 (coluna).
# Remover uma COLUNA não informativa.
df_clean_noduplicate_nouserid = df_clean_noduplicate.drop("user_id", axis=1)

# Remover uma LINHA.
df_clean_noduplicate.drop(5, axis=0)

```

```python
#Gráfico de identidade
plt.figure()

sns.kdeplot(df_clean_noduplicate_nouserid, x="age")

plt.show()
```

```python
plt.figure(figsize=(16,15))

sns.heatmap(
    df_clean_noduplicate_nouserid.corr(),
    annot = True,
    cmap="coolwarm"
)

plt.show()
```
